```python
class PizzaDelivery:

    def __init__(self, name, price, ingredients):
        self.name = name
        self.price = price
        self.ingredients = ingredients
        self.ordered = False

    def add_extra(self, ingredient, quantity, price_per_ingredient):
        if self.ordered:
            return f"Pizza {self.name} already prepared, and we can'nt make changes!"
        if ingredient in self.ingredients:
            self.ingredients[ingredient] += quantity
        else:
            self.ingredients[ingredient] = quantity
        self.price += price_per_ingredient * quantity

    def remove_ingredient(self, ingredient, quantity, price_per_ingredient):
        if self.ordered:
            return f"Pizza {self.name} already prepared, and we can'nt make changes."
        if ingredient not in self.ingredients:
            return f"Wrong ingredient selected! We do not use {ingredient} in {self.name}"
        if self.ingredients[ingredient] < quantity:
            return f"Please chack again the desired quantity of {ingredient}!"
        self.ingredients[ingredient] -= quantity
        self.price -= quantity * price_per_ingredient

    def make_order(self):
        if not self.ordered:
            self.ordered = True
            ingredient = ', '.join([f'{key} : {value}' for key, value in self.ingredients.items()])
            return f"Your ordered pizza {self.name} prepared with {ingredient} and the price will be {self.price}."
        return f"Pizza {self.name} already prepared, and we can't make any changes!"


Pepperoni = PizzaDelivery('Pepperoni', 499, {'cheese':2, 'Schezwan':4})
Pepperoni.add_extra('mozerella', 1,0.5)
Pepperoni.add_extra('cheese',1,1)
Pepperoni.remove_ingredient('cheese',1,1)
print(Pepperoni.remove_ingredient('bacon', 1, 2.5))
print(Pepperoni.remove_ingredient('tomatoes', 2, 0.5))
Pepperoni.remove_ingredient('cheese', 2, 1)
print(Pepperoni.make_order())
print(Pepperoni.add_extra('cheese', 1, 1))
```
