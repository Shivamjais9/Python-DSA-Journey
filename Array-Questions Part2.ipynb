**21. Find the sum of all elements using a while loop instead of a for loop.**

```python
num = ([10,20,30,40,50])
sum_of_num = 0
index = 0

while index < len(num):
    sum_of_num += num[index]
    index += 1

print(f'Sum of all numbers is: {sum_of_num}')
```

**22. Find the product of all elements in an array.**

```python
numbers2 = [2,3,4,5,6,7,8,9]
product_of_numbers = 1
index = 0

while index < len(numbers2):
    product_of_numbers *= numbers2[index]
    index += 1

print(f'Product of all numbers is: {product_of_numbers}')
```

**23. Find the difference between the largest and smallest element.**

```python
numbers = [11,22,33,44,55,66,77,88,99]

max_num = numbers[0]
min_num = numbers[0]

for i in numbers:
    if i > max_num:
        max_num = i
    if i < min_num:
        min_num = i

diff = max_num - min_num

print(f'Difference between largest and smallest number is: {diff}')
```

**24. Count how many elements are greater than the average.**

```python
numbers = [11,22,33,44,55,66,77,88,99]

sum_num = 0
for i in numbers:
    sum_num += i
    
avg_num = sum_num / len(numbers)

count = 0
for num in numbers:
    if num > avg_num:
        count += 1

print(f'No. of elements greater than the average: {count}')
```

**25. Move all zeros to the end of the array (maintain order of others).**

```python
# Solve by me
numbers = [0,22,34,21,0,54,27,0,98,75]

new_num = []
for num in numbers:
    if num == 0:
        new_num.append(num)

new_num2 = []
for num2 in numbers:
    if num2 != 0:
        new_num2.append(num2)

total = new_num2 + new_num 
print(total)
```
