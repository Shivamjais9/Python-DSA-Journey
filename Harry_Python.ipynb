**Take a number n from user input. Use a while loop to calculate the sum of numbers from 1 to n.**

```python
n = int(input("Enter a number: "))

i = 1          # starting number
total = 0      # variable to store sum

while i <= n:
    total += i   # add current number to total
    i += 1       # move to next number

print("Sum from 1 to", n, "is:", total)
```

**Problem 4: Write a recursive function to calculate the sum of first n natural numbers.**

```python
# By using loop function:
def sum_of_n(n):  
    sum1 = 0
    for i in range(1,n+1):
        sum1 += i
    print(sum1)

sum_of_n(5)
```

-  **But this is a loop function we wrote here**

- **By using recursion function**

  
**A recursive function is a function that:**

- Calls itself
- Has a base case (stopping condition)

```python
# By using recursion function:
def sum_n(n):
    if n == 1:
        return 1
    else:
        return n + sum_n(n-1)

sum_n(5)
```

**Why this is recursive?**

- It calls itself → sum_of_n(n - 1)
- It stops when n == 1 (base case)
- No loop is used

```python
def n_line(n):
    for i in range(n,0,-1):
        print('*'*i)

n_line(5)
```
