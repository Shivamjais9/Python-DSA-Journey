## 1️⃣ Student Class

Create an object of the class and print the student's details.

## 3️⃣ Rectangle Class
Create a class `Rectangle` with:
- attributes: length, width  
- method: `area()` to calculate and return the area  

Create an object and print the area.

---

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width

    def rectangle_area(self):
        area = self.length*self.width
        return f'Area of rectangle is {area}cm²'

Rec_area = Rectangle(4,6)
print(Rec_area.rectangle_area())
```

## 4️⃣ BankAccount Class
Create a class `BankAccount` with:
- attributes: account_holder, balance  
- method: `deposit(amount)`  
- method: `withdraw(amount)`  

Create an object and perform deposit and withdrawal operations.

---

```python
class BankAccount:
    def __init__(self, Name, balance):
        self.Name = Name
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        return f'Hello {self.Name} your {amount}Rs has been deposited, Current account balance is {self.balance}'

    def withdraw(self,amount):
        if self.balance >= amount:
            self.balance -= amount
            return f'Your {amount}Rs withdrawn. Remaining balance is: {self.balance}'
        else:
            return 'Amount is insufficient'
    
Customer = BankAccount('Shivam',0)
print(Customer.deposit(10000))
print(Customer.withdraw(5001))
```

## 5️⃣ Person Class
