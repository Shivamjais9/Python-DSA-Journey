**9.Count how many even numbers are present in an array.**

```python
even_num = []
for i in arr:
    if i % 2 == 0:
        even_num.append(i)

print(len(even_num))
```

**10. Count how many odd numbers are present in an array.**

```python
odd_num = []
for i in arr:
    if i % 2 != 0:
        odd_num.append(i)

print(len(odd_num))
```

**16. Remove duplicate elements from an array.**

```python
arr3 = np.array([2,3,2,44,3,6,7,44,7,23,2])
print(np.unique(arr3))
```

**17. Find the frequency of each element in an array.**

```python
arr3 = np.array([2,3,2,44,3,6,7,44,7,23,2])
(unique, counts) = np.unique(arr3, return_counts=True)
freq = dict(zip(unique, counts))
print(freq)
```

**18. Merge two arrays into one.**

**19. Sort an array in ascending order (without using built-in sort)**

```python
arr4 = np.array([5,2,9,1,7])
# Simple Bubble Sort
for i in range(len(arr4)):
    for j in range(0, len(arr4)-i-1):
        if arr4[j] > arr4[j+1]:
            arr4[j], arr4[j+1] = arr4[j+1], arr4[j]

print(arr4)
```

**20. Rotate an array to the right by 1 position**

```python
arr5 = np.array([1,2,3,4,5])
rotated = np.empty_like(arr5)
rotated[0] = arr5[-1]
rotated[1:] = arr5[:-1]
print(rotated)
```
