**Question 1**
- Array: [1, 2, 2, 3, 3, 3, 4]
- Print frequency of each element.

```python
arr = [1,2,2,3,3,3,4]

freq_dict = {}
for num in arr:
    count = 0
    for i in arr:
        if i == num:
            count += 1
    if num not in freq_dict:
        freq_dict[num] = count

print(f'Frequency of each elements is {freq_dict}')
```

**Question 3**
- Array: [10, 20, 10, 20, 30]
- Print frequency of only elements that appear more than once.

```python
arr = [10,20,10,20,30]

freq_dict = {}
for num in arr:
    count = 0
    for i in arr:
        if i == num:
            count += 1
    if num not in freq_dict and count > 1:
        freq_dict[num] = count

print(f'Frequency of element that appear more than once: {freq_dict}')
```

**Question 4**
- Array: [7, 8, 9, 7, 8, 7]
- Print frequency in the format: "element 7 occurs 3 times" (loop over array).

```python
arr = [7,8,9,7,8,7]
freq_dict = {}
for num in arr:
    count = 0
    for i in arr:
        if i == num:
            count += 1
    if num not in freq_dict:
        freq_dict[num] = count

for key in freq_dict:
    print(f'element {key} occurs {freq_dict[key]} times')
```

**Question 5**
- Array: [1, 2, 3, 4, 5, 1, 2, 1]
- Print the most frequent element and its frequency.

```python
arr = [1,2,3,4,5,1,2,1]

freq_dict = {}
for num in arr:
    count = 0
    for i in arr:
        if i == num:
            count += 1
    if num not in freq_dict:
        freq_dict[num] = count


max_value = -1
max_key = None

for key, value in freq_dict.items():
    if value > max_value:
        max_value = value
        max_key = key

print(f'Most frequent element with frequnecy: {max_key}:{max_value}')
```
